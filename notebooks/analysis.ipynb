```python
import librosa
import pretty_midi
import os

# フォルダ指定
data_dir = '../data'

# ファイル名の指定（実際の名前に書き換えてください）
audio_file = 'Bach_BWV849-01_001_20090916-SMD.mp3'
midi_file  = 'Bach_BWV849-01_001_20090916-SMD.mid'
```

```python
# 1. オーディオの読み込み
audio_path = os.path.join(data_dir, audio_file)
y, sr = librosa.load(audio_path)
print(f"Audio: {audio_file} を読み込みました (長さ: {len(y)/sr:.2f}秒)")
```

```python
# 2. MIDIの読み込み
midi_path = os.path.join(data_dir, midi_file)
pm = pretty_midi.PrettyMIDI(midi_path)
print(f"MIDI: {midi_file} を読み込みました")

# MIDIデータの情報を少し覗いてみる
print(f"MIDI内の楽器数: {len(pm.instruments)}")
for instrument in pm.instruments:
    print(f" - 楽器名: {instrument.name}, ノート数: {len(instrument.notes)}")
    
```
